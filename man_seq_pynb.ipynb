```python
input_multifasta = 'Contig1_genes (3).fasta'
output_tRNA_fasta = 'output_tRNA.fasta'

from Bio import SeqIO

def extrair_tRNA(input_file, output_file):
    with open(output_file, 'w') as out_file:
        for record in SeqIO.parse(input_file, 'fasta'):
            if record.description.startswith('tRNA'):
                SeqIO.write(record, out_file, 'fasta')

extrair_tRNA(input_multifasta, output_tRNA_fasta)
```

```python
output_file = 'output_tRNA.fasta'

with open(output_file, 'r') as file:
    content = file.read()
    print(content)
```

```python
input_multifasta = 'Contig1_genes (3).fasta'
output_rRNA_fasta = 'output_rRNA.fasta'

from Bio import SeqIO

def extrair_rRNA(input_file, output_file):
    with open(output_file, 'w') as out_file:
        for record in SeqIO.parse(input_file, 'fasta'):
           if 'RNA' in record.description and not record.description.startswith('tRNA'):
                SeqIO.write(record, out_file, 'fasta')

extrair_rRNA(input_multifasta, output_rRNA_fasta)
```

```python
output_file = 'output_rRNA.fasta'

with open(output_file, 'r') as file:
    content = file.read()
    print(content)
```

```python
from Bio import SeqIO

input_multifasta = 'Contig1_genes (3).fasta'
output_genes = 'output_genes.fasta'

def extrair_genes(input_file, output_file):
    with open(output_file, 'w') as out_file:
        for record in SeqIO.parse(input_file, 'fasta'):
            # Verifica se 'RNA' não está presente na descrição e não é uma sequência de tRNA
            if 'RNA' not in record.description and not record.description.startswith('>tRNA'):
                SeqIO.write(record, out_file, 'fasta')

extrair_genes(input_multifasta, output_genes)
```

```python
output_file = 'output_genes.fasta'

with open(output_file, 'r') as file:
    content = file.read()
    print(content)
```
